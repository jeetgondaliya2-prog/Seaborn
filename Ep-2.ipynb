```python
import seaborn as sns                        # categorical plot
import matplotlib.pyplot as pt
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.countplot(x=df['sex'])          # count no of male and female
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])    
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'])                     # in barplot x-categorical and y-quantitaive (default plot is calculated average) 
```

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)          # total sum of bill for men and women
```

```python
sns.boxplot(x='tip',y='day',data=df)              # boxplot
```

```python
sns.violinplot(x='day',y='total_bill',data=df)                       #violinplot
```

```python
  sns.stripplot(x='day',y='total_bill',data=df)                     #stripplot
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)                 #swarmplot
```

```python
sns.violinplot(x='day',y='total_bill',data=df)       #combination of violinplot and swarmplot
sns.swarmplot(x='day',y='total_bill',data=df)
```
